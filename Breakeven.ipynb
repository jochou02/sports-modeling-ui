```python
import matplotlib.pyplot as plt
import numpy as np
from math import comb
plt.style.use('seaborn-whitegrid')
```

```python
# @param start: x axis interval starting value
# @param end: x axis interval ending value
# @param save: saves graph locally if set to True
# @param id: ID of the graph for saving
def graph(start, end, save=False, id=0):
    
    fig = plt.figure()
    ax = plt.axes()
    fig.set_size_inches(20, 10)

    x = np.linspace(start, end, 100)

    y = 3 * x**2
    plt.plot(x, y, label='2 Power')

    y = 5 * x**3
    plt.plot(x, y, label='3 Power')

    y = 2.25 * x**3 + 1.25 * comb(3,2) * (x**2 * (1-x))
    plt.plot(x, y, label='3 Flex')

    y = 10 * x**4
    plt.plot(x, y, label='4 Power')

    y = 5 * x**4 + 1.5 * comb(4, 3) * (x**3 * (1-x))
    plt.plot(x, y, label='4 Flex')

    y = 10 * x**5 + 2 * comb(5, 4) * (x**4 * (1-x)) + 0.4 * comb(5, 3) * (x**3 * (1-x)**2)
    plt.plot(x, y, label='5 Flex')

    y = 25 * x**6 + 2 * comb(6, 5) * (x**5 * (1-x)) + 0.4 * comb(6, 4) * (x**4 * (1-x)**2)
    plt.plot(x, y, label='6 Flex')

    y = 30 * x**6 + 2 * comb(6, 5) * (x**5 * (1-x)) + 0.4 * comb(6, 4) * (x**4 * (1-x)**2)
    plt.plot(x, y, label='6 Flex (30x)')

    y = 1 * x**0
    plt.plot(x, y, label='EV = 1', color='black')

    plt.title('Graph of Expected Values')
    plt.xlabel('Probability of hitting each pick')
    plt.ylabel('Expected Return')
    plt.legend(loc='upper left')
    plt.grid()
    plt.show()

    if save:
        fig.savefig(f'EV{id}.png', dpi=100)
```

```python
graph(0.5, 1, False, 10)
```

```python
graph(0.5, 0.6, False, 10)
```

```python
graph(0.52, 0.62, True, 11)
```
